# matplotlib

```python
import matplotlib.pyplot as plt
```

```python
a = [2,4,6]   #데이터의 배열을 말하는 a
b = [1,2,3]
plt.plot(b,a, ls = ':')  #b를 x축으로 a를 y축으로 설정 #line style = ls
plt.show()
```

```python
a = [2,4,6]   
b = [1,2,3]
plt.plot(b,a, ls = ':', marker = 'o') #두 데이터 참조하여 그리는 것으로, 그 두 지점이 만나는    # 주요 데이터 지점에 마커 표시
plt.show()
```

```python
a = [2,4,6]   
b = [1,2,3]
plt.plot(b,a,lw = 5) #line width  
plt.show()
```

```python
a = [2,4,6]   
b = [1,2,3]
plt.plot(b,a, color = 'green') #line width  
plt.show()
```

```python
a = [2,4,6]   
b = [1,2,3]
plt.plot(b,a, ls = '--', lw = 2, color = 'purple', marker = 'o', markerfacecolor = 'r', markeredgecolor = 'black', markersize = '10')
#markerfacecolor 마커의 표면색 설정
#markeredgecolor 마커의 테두리색 설정
#markersize  마커의 크기 설정
#markeredgewidth 마커의 테두리선 두께
plt.show()
```

```python
import numpy as np

a = np.arange(7)
b = [1,4,5,8,9,5,3]
plt.plot(a, b, ls = '--', color = 'green', marker = 'o')
plt.show()
```

```python
a = np.arange(7)
b = [1,4,5,8,9,5,3]
plt.plot(a, b, ls = ':', color = 'red', marker = 'd', markerfacecolor = 'blue', markersize = '10', markeredgewidth = '0.8')
plt.show()
```

```python
a = np.arange(7)
b = [1,4,5,8,9,5,3]
plt.plot(a, b, ls = ':', color = 'red', marker = 'd', markerfacecolor = 'blue', markersize = '10', markeredgewidth = '0.8')

plt.xlim(-2,10) #각 축의 눈금 범위 설정
plt.ylim(-5, 12)

plt.show()
```

```python
a = np.arange(7)
b = [1,4,5,8,9,5,3]
plt.plot(a, b, ls = ':', color = 'red', marker = 'd', markerfacecolor = 'blue', markersize = '10', markeredgewidth = '0.8')
plt.xlim(-1,8)
plt.ylim(0,10)

plt.xticks([0,1,2,3,4, 5,6,7,8,])  #각 축의 눈금 표시 설정
plt.yticks([1,2,3,4,5,6,7,8,9])

plt.grid()           #차트 전체의 그리드 표시 (ticks로 설정했던것과 연동이 되어 나타남)
plt.show()
```

```python
df = pd.read_excel('Data/시도별 전출입 인구수.xlsx', fillna = 0, header = 0)
```

```python
# 전출지별에서 누락값을 앞 데이터로 채움
df = df.fillna(method = 'ffill')
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask = (df['전출지별']=='서울특별시') & (df['전입지별']!='서울특별시')   #1,0에 해당하는 값이 들어있는 mask
df_seoul = df[mask]
df_seoul
```

```python
df_seoul = df_seoul.drop(['전출지별'], axis = 1) #axis=1은 열의 방향
df_seoul.rename({'전입지별':'전입지'}, axis = 1, inplace = True)
df_seoul.set_index('전입지',inplace = True)
df_seoul
```

```python
# 서울에서 다른 지역으로 이동한 (경기도로 이동한) 데이터만 추출하여 정리
sr_one = df_seoul.loc['경기도']
sr_one  #두개가 하나로 묶어져있는 딕셔너리 타입
```

```python
plt.plot(sr_one.index, sr_one.values)
```

```python
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.title('서울->경기도 이동한 인구')
plt.plot(sr_one.index, sr_one.values)
plt.show()
```

```python
col_years = list(map(str, range(1970, 2018)))
df_seoul
```

```python
df_4 = df_seoul.loc[['충청남도','경상남도','강원도','전라남도'], col_years]
df_4 = df_4.transpose()
```

```python
plt.style.use('ggplot')
```

```python
df_4.index = df_4.index.map(int)
```

```python
df_4.plot(kind = 'area', stacked = True, alpha = 0.3, figsize=(20,10))  
#stacked옵션 -> True일때 데이터가 누적되어((쌓여서) 함께 나옴(막대그래프는 막대그래프 위에 쌓여서 출력됨)
#alpha = 투명도(1.0이 원래)
plt.show()
```

```python
df_4.plot(kind ='bar', figsize=(20,10), width = 0.7, color = ['orange','green','skyblue', 'blue'])
#width => bar의 =width값
```
