```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#타이타닉 데이터셋 로드
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.set_style('darkgrid')               #차트 스타일 설정
fig = plt.figure(figsize = (15,5))      #전체그림을 뜻하는 figure (여러 x가 들어갈 수 있음)
ax1 = fig.add_subplot(1,2,1)  #1행2열의 첫번째 객체
ax2 = fig.add_subplot(1,2,2)  #1행2열의 두번째 객체

#ax1 = fig.add_subplot(1,3,1) 
#ax2 = fig.add_subplot(1,3,2) 
#ax3 = fig.add_subplot(1,3,3)


sns.regplot(x='age',            # X축 변수
            y='fare',           # Y축 변수
            data=titanic,       # 데이터 셋
            ax=ax1)             # axe 객체

sns.regplot(x='age',            # X축 변수
            y='fare',           # Y축 변수
            data=titanic,       # 데이터 셋
            ax=ax2,             # axe 객체
           fit_reg = False      #회귀선 미표시(오른쪽 차트)
           )   
# sns.regplot(x='age',            # X축 변수
#             y='fare',           # Y축 변수
#             data=titanic,       # 데이터 셋
#             ax=ax3,             # axe 객체
#            fit_reg = False      #회귀선 미표시(오른쪽 차트)
#            )   

plt.show()
```

```python
sns.set_style('darkgrid')               #차트 스타일 설정
fig = plt.figure(figsize = (20,5))      #전체그림을 뜻하는 figure (여러 x가 들어갈 수 있음)
ax1 = fig.add_subplot(1,3,1)  #1행2열의 첫번째 객체
ax2 = fig.add_subplot(1,3,2)  #1행2열의 두번째 객체
ax3 = fig.add_subplot(1,3,3)

sns.distplot(titanic['fare'], ax = ax1)
sns.distplot(titanic['fare'], hist = False, ax = ax2)
sns.distplot(titanic['fare'], kde = False, ax = ax3)
```

```python
table = titanic.pivot_table(index = ['sex'], columns = ['class'], aggfunc = 'size')
sns. heatmap(table,
            annot = True, fmt = 'd',
            cmap = 'YlGnBu',
            linewidth = 5,
            cbar = False)
plt.show()
```

```python
sns.set_style('whitegrid')

fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x='class',
              y='age',
              data= titanic,
              ax = ax1 )

sns.stripplot(x='class',
              y='age',
              data= titanic,
              ax = ax2 )

plt.show()
#속성별 y축에분포에 맞춰서 보여줌
```
