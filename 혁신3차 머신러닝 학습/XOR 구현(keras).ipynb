### X data

```python
X = np.array([[0,0],
             [1,0],
             [0,1],
             [1,1]])
```

### y data

```python
y = np.array([0,1,1,0])
```

### Dense()

Dense 클래스를 이용하여 레이어의 노드를 만드는데, Sequentioal 모델에 추가하면 다른 노드들과 연결된다

즉, 입출력 레이어를 모두 연결해주는 것이 Dense Layer이다

​

모델을 추가하는 방식에 대해서 조금 더 설명을 한다

model.add(Dense(2, input_dim=2)) : 레이어의 수를 2개로 두고, 그 노드에 입력 값을 2개로 사용한다는 것

model.add(Dense(5, input_dim=3)) : 레이어의 수를 5개로 두고, 그 노드에 입력 값을 3개로 사용한다는 것

model.add(Dense(1)) : 레이어의 수를 1개로 두는데, 입력 값으로는 이전 층의 결과들을 사용한다는 것

model2.add(Dense(2, input_dim=2)) : 이렇게 두개를 하면 아래 그림처럼 된다

```python
# 모델생성
model = Sequential()
model2 = Sequential()
model.add(Dense(30, input_dim=2, activation='sigmoid'))   
model.add(Dense(20, activation='sigmoid'))
model.add(Dense(4, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
# 먼저 위에꺼부터 할것
model2.add(Dense(8, input_dim=2, activation='sigmoid'))
model2.add(Dense(16, activation='sigmoid'))
model2.add(Dense(32, activation='sigmoid'))
model2.add(Dense(64, activation='sigmoid'))
model2.add(Dense(128, activation='sigmoid'))
model2.add(Dense(64, activation='sigmoid'))
model2.add(Dense(32, activation='sigmoid'))
model2.add(Dense(16, activation='sigmoid'))
model2.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일
model.compile(loss="binary_crossentropy", 
              optimizer='adam',
             metrics=['accuracy'])

model2.compile(loss="binary_crossentropy", 
              optimizer='adam',
             metrics=['accuracy'])
```

```python
# 모델 학습
model.fit(X,y,epochs=5000)
model2.fit(X,y,epochs=5000)
```

```python
model.predict(X)
model2.predict(X)
```

```python
model.evaluate(X,y)
```

### 총정리

1. 퍼셉트론 : y = wx +b + activation function

2. 다층퍼셉트론 (MLP)

3. 활성화 함수는 비선형 함수
  - step 함수 : 사용하지 않는다.
  - sigmoid 함수 : 사용한다.

4. 문제 유형에 따라 출력층의 활성화 함수, 오차함수가 다르다.
  - 회귀 : linear / mse
  - 이진분류 : sigmoid /binary_crossentropy
  - 다중분류 : softmax / categorical_corssentropy
  
5. Sequential / Dense
  - Sequential : 딥러닝 모델의 틀
  - Dense : layer(층)
